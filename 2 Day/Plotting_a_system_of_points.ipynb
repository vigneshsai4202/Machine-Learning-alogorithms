```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,40,1000) # starting point,finishing point ,no of points
```

Distance travelled by the Robber=2.5t

```python
d_r=2.5*t
```

Distance travlled by the shref d=3(t-5)

```python
d_s=3*(t-5)
```

```python
fig, ax = plt.subplots()
plt.title('A Bank Robber Caught')
plt.xlabel('Time (in minutes)')
plt.ylabel('Distance (in km)')
ax.set_xlim([0, 40])
ax.set_ylim([0,100])
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```
